# Regresiones lineales


## Introducción

Es una técnica para identificar relaciones entre una variable dependiente y una o varias variables independientes. Esta técnica es utilizada para hallar efectos de causalidad entre las variables. 

La ventaja de realizar un analisis de regresion lineal es que permite identificar tanto la relacion entre variables como la fortaleza de la relacion de la(s) variable(s) explicativas.

La regresión lineal proviene de la estadistica inferencial en la cual su proposito es producir **Modelos para generar datos**

La inferencia consta de 3 pasos: 

1. Estimación de Parametros
2. Prueba del modelo
3. Pruebas de hipotesis

En este cuaderno de Jupyter veremos entonces las Regresiones Lineales Simples usando R.

```r
library(tidyverse)
```

La Regresión Lineal se expresa como:

## $y=\beta_0+\beta_1x+\epsilon$

Donde:

$\beta_0$: Corresponde al intercepto de la recta con el eje y

$\beta_1$: Corresponde a la pendiente de la recta, que refleja cuánto varía ala respuesta cuando aumentamos una unidad de x

$\epsilon$:

>1. Son errores aleatorios incorrelacionados, 
2. con media cero y 
3. varianza constante $\sigma^2$



## Estimación de la recta de regresión Mínimos Cuadrados Ordinarios (OLS)

<img src="Regresion_lineal.png",width=600,height=600>

Se obtienen los valores de $\beta_i$ minimizando las distancias entre los puntos observados y los predichos por la recta de ajuste como se observa en la figura

# Ejemplo

```r
#Ejemplo

x<- rnorm(1000, mean = 1, sd = 1)  #Genera 1000 datos aleatorios para x
y<- rnorm(1000, mean = 1, sd = 2)  #Genera 1000 datos aleatorios para y

modelo <- lm(y~x)                  #genera un modelo de y en funcion de x

modelo                             #Imprime el resultado

cat('ahora veamos los resultados del modelo!!!' )

summary(modelo)                    #Resumen estadistico del modelo
```

## Estimación por Máxima Verosimilitud:

Habitualmente el objetivo de un análisis de regresión no consiste únicamente en estimar la recta, sino en _inferir_ con ella, asociar un error a las estimaciones obtenidas, contrastar un determinado valor de los parámetros, o predecir la respuesta para un valor dado junto con una banda de confianza. Por tanto nos apoyamos en distribuciones de probabilidades de la variable de respuesta o lo mismo sobre una distribución sobre los errores

### $\epsilon_i \sim N(0,\sigma^2)$

es decir:

>1. Errores independientes e identicamente distribuidos
2. De acuerdo con una distribución Normal de media $\mu=0$
3. Varianza constante $\sigma^2$

Por lo anterior la solución más factible a la vista de los datos observados equivale a obtener la solución de máxima-verosímilitud

**En regresión lineal la solución de OLS coincide con la máxima-verosimilitud**




```r
options(repr.plot.width=10, repr.plot.height=5)
plot(modelo$fitted.values,type='l',col='black')
lines(c(rep(NA,1000),rnorm(100,1.02,0.01)),col='red')
```

## Inferencia sobre $\hat\beta_0$ y $\hat\beta_1$

Una vez estimados los valores de $\hat\beta_0$ y $\hat\beta_1$ se puede determinar su distribución de probabilidades utilizando el estadístico _t_.

Los intervalos de confianza permiten determinar si los valores hallados son insesgados o no bajo la siguiente prueba:


$H_0: \beta_i = \beta^*$ 

$H_1: \beta_i \not = \beta^*$

Debemos validar si los estimadores son significativos para la regresión, entonces es:

$H_0: \beta_i = 0$ 

$H_1: \beta_i \not = 0$


Alternativamente se rechaza $H_0$ a un nivel $\alpha$ cuando $p-value \leq \alpha$

Si conseguimos rechazar esta hipotesis se puede concluir que la variable $X$ tiene relación con la variable $y$.

# Siguiente paso

Hasta este momento hemos logrado determinar que el modelo ajustado ha obtenido los regresores adecuados para modelar el comportamiento de la variable explicada, veamos ahora cómo es el modelo para pronosticar

## Validación de errores

```r
options(repr.plot.width=10, repr.plot.height=7)
par(mfrow=c(2,2))
plot(modelo$residuals,type='p',main='Gráfico de Dispersión',pch=16,cex=0.8) # Grafica de dispersion de los errores
hist(modelo$residuals,freq = FALSE, main = 'Histograma')                    #Histograma y Distribucion empirica
lines(density(modelo$residuals),col = 'red',lwd =2)                         #
qqnorm(modelo$residuals)                                                    #Grafico QQ      
qqline(modelo$residuals)                                                    #
acf(modelo$residuals, main = 'acf')                                         # Autocorrelograma de los errores
```

# Pruebas estadisticas de normalidad de los errores


```r
shapiro.test(modelo$residuals)
```
